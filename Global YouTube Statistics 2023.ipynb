# Visualization

```python
# There is a strong correlation between video views and subscribers
you1['video views'].corr(you1['subscribers'])
```

```python
you1.plot.scatter(x='video views', y='subscribers')
plt.xlabel('Video Views')
plt.ylabel('Subscribers')
plt.title('Relationship Between Subscribers and Video Views')
```

```python
# There is a weak correlation between video views and monthly earnings.
# This because other factors affect the monthly earnings
# Some of them are: 1. video lenght 2. Ad blockers 3. Ad quality 4. number of clicks and Ad recieves

you1['highest_monthly_earnings'].corr(you1['video views'])
```

```python
you1.plot.scatter(x='video views', y='highest_monthly_earnings')
plt.xlabel('Video Views')
plt.ylabel('Highest Monthly Earnings')
plt.title('Relationship Between Video Views and Highest Monthly Earnings')
```

```python
you1['highest_monthly_earnings'].corr(you1['subscribers'])
```

```python
you1.plot.scatter(x='subscribers', y='highest_monthly_earnings')
plt.xlabel('Subscribers')
plt.ylabel('Highest Monthly Earnings')
plt.title('Relationship Between Subscriberss and Highest Monthly Earnings')
```

# Highest Earning Categories

```python
highest_earning_category = you1.groupby(['category'])['highest_monthly_earnings'].sum()
highest_earning_category
```

```python
highest_earning_category.plot.bar()
plt.xlabel('category')
plt.ylabel('highest_monthly_earnings')
plt.title('Highest Monthly Earnings by Category')
```

# Categories with the highest number of subscribers

```python
category_with_highest_subscribers = you1.groupby(['category'])['subscribers'].sum()
category_with_highest_subscribers
```

```python
category_with_highest_subscribers.plot.bar()
plt.xlabel('category')
plt.ylabel('Subscribers')
plt.title('Highest Subscribers by Category')
```

# How countries and their population contribute to the success of youtube channels

```python
highest_subscribers_by_population = you1.groupby(['Country'])['Population'].sum()
highest_subscribers_by_population
```

```python
highest_subscribers_by_population.plot.bar(figsize=(15, 7))
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population by Country')
```

# Does unemployment rate play a role also?

```python
Unemployment_rate_by_country = you1.groupby(['Country'])['Unemployment rate'].sum()
Unemployment_rate_by_country
```

```python
Unemployment_rate_by_country.plot.bar(figsize=(15, 7))
plt.ylabel('Unemployment rate')
plt.xlabel('Country') 
plt.title('Unemployment_rate_by_country') 
```

# Bigger population means bigger youtube users 

```python
Subscribers_by_country = you1.groupby(['Country'])['subscribers'].sum()
Subscribers_by_country
```

```python
Subscribers_by_country.plot.bar(figsize=(15, 7))
plt.xlabel('Country')
plt.ylabel('Subscribers')
plt.title('Subscribers by Country')
```

# And bigger youtube users means bigger monthly earnings for youtubers

```python
Subscribers_by_country = you1.groupby(['Country'])['highest_monthly_earnings'].sum()
Subscribers_by_country
```

```python
Subscribers_by_country.plot.bar(figsize=(15, 7))
plt.xlabel('Country')
plt.ylabel('Monthly earnings')
plt.title('Monthly Earnings by Country')
```
